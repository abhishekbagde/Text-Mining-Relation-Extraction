```python
import json

# Function to load JSON data
def load_json_data(filepath):
    with open(filepath, 'r', encoding='utf-8') as file:
        data = json.load(file)
    return data

# Load data
train_data = load_json_data('DocRED-master/data/train_annotated.json')
rel_info = load_json_data('DocRED-master/data/rel_info.json')
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

def find_relations(text, rel_info):
    # Tokenize and tag the sentence
    tokens = word_tokenize(text)
    tagged = pos_tag(tokens)
    
    # Here you would implement your rule-based logic to extract relations
    # This might involve looking for specific sequences of POS tags
    # or specific named entities followed by verbs or prepositions indicating a relation
    
    # Example: finding a simple noun-verb-noun pattern

    relations = []
    for i in range(len(tagged)-2):
        if tagged[i][1] == 'NNP' and 'country' in tagged[i][0]:
            if i > 0 and tagged[i-1][1] == 'IN':
                relation = f"{tagged[i-1][0]} {tagged[i][0]}"
                relations.append(relation)
    return relations
```

```python
for item in train_data:
    # Assuming 'sents' contains the sentences to process
    for sent in item['sents']:
        sentence = " ".join(sent)  # Join tokens to form a string sentence, if necessary
        relations = find_relations(sentence, rel_info)
        print(relations)
```
